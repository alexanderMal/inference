```python
# Manually created Graph to create DAG from data learned structure
from pgmpy.models import BayesianNetwork
import os
os.environ['NUMEXPR_MAX_THREADS'] = '40'


path = os.getcwd()+"/local/uni/inference/Task5/"


import matplotlib.pyplot as plt

model = BayesianNetwork([('U1', 'Education System'),
('U2', 'Academic Performance'),
('Grading System', 'Academic Performance'),
('Exam', 'Academic Performance'),
('Subject', 'Exam'),
('Learned Material', 'Academic Performance'),
('Institute', 'Education System'),
('Teacher', 'Education System'),
('Teacher', 'Student Motivation'),
('U3', 'Student Motivation'),
('Student', 'Student Motivation'),
('Student Motivation', 'Time Spend Learning'),
('Parental Supervision', 'Time Spend Learning'),
('Student Motivation', 'Learning Groups'),
('Learning Groups', 'Time Spend Learning'),
('Time Spend Learning', 'Learned Material'),
('Student', 'Affinity'),
('Subject', 'Affinity'),
('Affinity', 'Previous Performance'),
('Affinity', 'Learned Material'),
('Education System', 'Learned Material'),
('Education System', 'Student Motivation'),
('U4', 'Previous Performance'),
('Education System', 'Academic Performance'),

])
viz = model.to_graphviz()

viz.draw(path + "./dag.png", prog="dot")
#show image from path in plt
img = plt.imread(path + "./dag.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
from pgmpy.models import BayesianNetwork
import os
os.environ['NUMEXPR_MAX_THREADS'] = '40'


path = os.getcwd()+"/local/uni/inference/Task6/"


import matplotlib.pyplot as plt

model = BayesianNetwork([('Health Condition', 'Exercise'),
                         ('Health Condition', 'Diet'),
                         ('Health Condition', 'Weightloss'),
                         ('Diet', 'Weightloss'),
                         ('Exercise', 'Weightloss'),
                         ('Motivation', 'Diet'),
                         ('Motivation', 'Exercise'),

])
viz = model.to_graphviz()

viz.draw(path + "./dag.png", prog="dot")
#show image from path in plt
img = plt.imread(path + "./dag.png")
plt.imshow(img)
plt.axis('off')
plt.show()
```
